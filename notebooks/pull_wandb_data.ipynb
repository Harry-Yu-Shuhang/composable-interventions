```python
from datasets import Dataset, DatasetDict
from tqdm import tqdm
import pandas as pd
import json
import os

import wandb
api = wandb.Api()
```

```python
project_name = "kyledevinobrien1/AK_Tests"
runs = api.runs(project_name)
for run in runs:
    if run.state != "finished":
        continue

    print("Run ID: ", run.id)
    print("Config")
    display(pd.read_json(json.dumps(run.config), orient="index").T)
    print("Summary Metrics")
    display(pd.read_json(json.dumps(run.summary_metrics), orient="index").T)
    break
```
